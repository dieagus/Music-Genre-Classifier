```python
import pandas as pd 
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
df=pd.read_csv('gdrive/My Drive/lyrics.csv')
```

```python
import re
nan_value = float("NaN")
#replaces "not available" with a none value then deletes all none values

# DATA PROCESSING

df.replace("Not Available", nan_value, inplace=True)
df.replace("Other", nan_value, inplace=True)
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["genre"], inplace=True)
df.dropna(subset = ["song"], inplace=True)
df.dropna(subset = ["lyrics"], inplace=True)
df.dropna(subset = ["year"], inplace=True)
df.dropna(subset = ["artist"], inplace=True)

df = df.replace(['\n'],'', regex=True) 
#df["lyrics"] = df.lyrics.str.replace('[^a-zA-Z0-9]', '')

#function for cleaning up all non alphabetical characters
def clean(text):
  text = re.sub("'", "", text)
  text = re.sub("(\\W)+", " ",text)
  return text

df["lyrics"] = df.lyrics.apply(clean)

```

```python
#instead of all lyrics this helps us extract the first 200 words of the song to help increase the speed of the model. all lyrcis would take too long
def get_split(text1):
  l_total = []
  l_parcial = []
  if len(text1.split())//150 >0:
    n = len(text1.split())//150
  else: 
    n = 1
  for w in range(n):
    if w == 0:
      l_parcial = text1.split()[:200]
      l_total.append(" ".join(l_parcial))
    else:
      l_parcial = text1.split()[w*150:w*150 + 200]
      l_total.append(" ".join(l_parcial))
  return l_total
train_df["get_split"] = train_df["lyrics"].apply(get_split)
```
